## 1.3 Prediction


1. Under the naive Bayes model and your estimates of the above probabilities, what is the most likely label our test example, $x_{new}$: "positive" or "negative"? Show your calculations.

$$x_{new} = \begin{bmatrix}1 & 1 & 0 & 1\end{bmatrix}$$

> Calculate the probabilities by hand.

$$p(\text{positive} \mid predictable = 1, fun = 1, pathetic = 0, satire = 1) = ?$$

$$p(\text{negative} \mid predictable = 1, fun = 1, pathetic = 0, satire = 1) = ?$$

### Answer:

$p(\text{positive} \mid predictable = 1, fun = 1, pathetic = 0, satire = 1) = $

$p(\text{positive}) * p(predictable = 1 \mid \text{positive}) * p(fun = 1 \mid \text{positive}) * p(pathetic = 0 \mid \text{positive}) * p(satire = 1 \mid \text{positive})$

$=(\frac{4}{10})(\frac{2}{4})(\frac{3}{4})(1)(\frac{1}{4})$

$=\frac{3}{80}$


$p(\text{negative} \mid predictable = 1, fun = 1, pathetic = 0, satire = 1) = $

$p(\text{negative}) * p(predictable = 1 \mid \text{negative}) * p(fun = 1 \mid \text{negative}) * p(pathetic = 0 \mid \text{negative}) * p(satire = 1 \mid \text{negative})$

$=(\frac{6}{10})(\frac{4}{6})(\frac{1}{6})(\frac{1}{6})(\frac{3}{6})$

$=\frac{1}{180}$

$p(x) \propto p(\text{positive} \mid x) + p(\text{negative} \mid x) = \frac{3}{80} + \frac{1}{180} = \frac{31}{720}$

$p(\text{positive} \mid x) = \frac{\frac{3}{80}}{\frac{31}{720}} = \frac{27}{31} \approx 0.87096 $

$p(\text{negative} \mid x) =  \frac{\frac{1}{180}}{\frac{31}{720}} = \frac{4}{31} \approx 0.12903$

For the test example, $x_{new}$, the most likely label is positive.

# QUESTION 2: WRITING A NAIVE BAYES CLASSIFIER FROM SCRATCH

Write a class to perform discrete Naive Bayes for a random dataset with n = 1000 (# of data points) and d = 2 (# of dimensions) for a binary classification y [0,1]. x1 and x2 can get any values between [0 and 1]. Then, implement the Laplace Smoothing function for the Naive Bayes class

### Import the necessary libraries

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tqdm import tqdm
from scipy.stats import norm
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
```

### Creating a sample dataset with two dimensions x1, x2

```python
x1 = np.random.randint(2, size = 1000)
x2 = np.random.randint(2, size = 1000)

y = np.random.randint(2, size = 1000)
```

### Create an array to store the discrete generated variables x1 and x2

```python
X = np.array(list(zip(x1, x2)))
```

### Create the class for Naive Bayes Classifier

```python
class Naive_Bayes:

    def __init__(self, X, y, laplace=0):
        self.X = X
        self.y = y
        self.laplace = laplace
        
        # Train-test split
        self.X_train = None
        self.X_test = None
        self.y_train = None
        self.y_test = None
        
        # Priors
        self.prior_0 = None
        self.prior_1 = None
        
        # Likelihoods
        self.x1_y0 = None
        self.x1_y1 = None
        self.x2_y0 = None
        self.x2_y1 = None

    def trainTestSplit(self):
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, self.y, test_size=0.2, random_state=42)

    def probability(self, X, prior, dist_1, dist_2):
        # Generating the probablity distributions for x1 and x2
        probability = prior
        for i in range(X.shape[0]):
            if X[i] == 1:
                probability *= dist_1[i]
            else:
                probability *= dist_2[i]
        return probability

    def runModel(self):
        # Creating subsets of training data for class 0 and class 1 (y) respectively, to use for likelihood calculation
        x_train_y0 = self.X_train[self.y_train == 0]
        x_train_y1 = self.X_train[self.y_train == 1]

        # Calculating the prior probablities of the class labels (y - 0 and 1)
        n0 = x_train_y0.shape[0]
        n1 = x_train_y1.shape[0]
        size = self.X_train.shape[0]
        
        self.prior_0 = n0 / size
        self.prior_1 = n1 / size

        # Let 'k' be the number of features in the dataset
        k = 2
        # YOUR CODE HERE (to include laplace smoothing)
        laplace = self.laplace

        # Calculating the likelihoods of the features x1 and x2, given class 0 and 1 respectively
        # Probability of x1 being 1 given y
        x1_y0_prob = np.sum(x_train_y0[:, 0] == 1)
        x1_y1_prob = np.sum(x_train_y1[:, 0] == 1)
        # Probability of x2 being 1 given y
        x2_y0_prob = np.sum(x_train_y0[:, 1] == 1)
        x2_y1_prob = np.sum(x_train_y1[:, 1] == 1)
        
        # Apply laplace smoothing
        self.x1_y0 = (x1_y0_prob + laplace) / (n0 + k * laplace)
        self.x1_y1 = (x1_y1_prob + laplace) / (n1 + k * laplace)
        self.x2_y0 = (x2_y0_prob + laplace) / (n0 + k * laplace)
        self.x2_y1 = (x2_y1_prob + laplace) / (n1 + k * laplace)

    def predict(self):
        predictions = []
        # Calculating the posterior probabilities for a sample being a part of each class label
        for idx in range(self.X_test.shape[0]):
            prob_y0 = self.probability(self.X_test[idx], self.prior_0, [self.x1_y0, self.x2_y0], [1 - self.x1_y0, 1 - self.x2_y0])
            prob_y1 = self.probability(self.X_test[idx], self.prior_1, [self.x1_y1, self.x2_y1], [1 - self.x1_y1, 1 - self.x2_y1])
            # Predicting the class label with the higher posterior probability
            if prob_y0 > prob_y1:
                predictions.append(0)
            else:
                predictions.append(1)
        return predictions
            
```

### Execute without Laplace Smooting

```python
nb = Naive_Bayes(X, y)
```

```python
# Added cell to train model
nb.trainTestSplit()
```

```python
nb.runModel()
```

```python
predictions = nb.predict()
```

```python
accuracy = accuracy_score(nb.y_test, predictions)
print("Accuracy of the Naive Bayes model:", accuracy)
```
